```python
import pymysql
import pandas as pd
import os


dbcon = pymysql.connect(host = "127.0.0.1",port = 3308 ,user= "root", passwd="my-secret-pw",db='tc_store_new', charset='utf8')
```

```python
SQL_Query = pd.read_sql_query(
        '''SELECT transaction_id, product_name, quantity
            FROM tc_store_new.order o
                JOIN tc_store_new.product p
                    ON o.product_id = p.product_ID
                WHERE
                    o.created_on > '2022-03-03';''', dbcon)
```

```python
SQL_Query
```

```python
all_transaction = []
for i, order in SQL_Query.iterrows():
    count = int(order['quantity'])
    each_order = (order['transaction_id'], order['product_name'])*count
    all_transaction.append(each_order)
```
